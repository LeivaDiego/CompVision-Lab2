```python
from skimage.morphology import dilation, erosion, disk
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

## Inciso 4
Obtener el gradiente morfológico 
$$ 
\nabla (I) = (I \oplus B) - (I \ominus B) 
$$ 

de la imagen `brain-scan.jpeg`.
Explicar cuál es el resultado obtenido y explicar por qué se le llama gradiente. (Sugerencia: experimentar con otras im´agenes
en escala de grises para entender el resultado de esta operaci´on)

```python
image = imread("images/brain-scan.jpeg", as_gray=True)

# Definir el elemento estructurante 
selem = disk(3)

dilated = dilation(image, selem)
eroded = erosion(image, selem)
morph_gradient = dilated - eroded

fig, axes = plt.subplots(1, 4, figsize=(15, 5))
titles = ['Imagen Original', 'Dilatación', 'Erosión', 'Gradiente Morfológico']
images = [image, dilated, eroded, morph_gradient]

for ax, img, title in zip(axes, images + [morph_gradient], titles):
    ax.imshow(img, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

plt.show()
```
